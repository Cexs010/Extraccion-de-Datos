```python
#librerias a usar
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#cargar archivo xlsx
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
```

```python
#Verificamos información del DataFrame
df.info()
```

**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtramos por CVE_DOC, que es el identificador unico de cada factura, en este caso se filtra por la factura F37105 y se crea un nuevo archivo csv.
filtro1 = df[(df["CVE_DOC"] == "F37105")]
filtro1.to_csv("filtroPorObjeto.csv")
filtro1
```

```python
#Filtro por filas no consecutivas, en este caso se filtra por las filas 1 y 2 y se crea un nuevo archivo csv.
filtro2= df.loc[[1,2],:] #Filas no consecutivas
filtro2.to_csv("filtroPorFilas.csv")
filtro2
```

```python
#Filtro por columnas consecutivas, en este caso se filtra por las columnas 1, 2 y 3 y se crea un nuevo archivo csv.
filtro3 = df.iloc[:, [1,2,3]] #Columnas consecutivas
filtro3.to_csv("filtroPorColumnas.csv")
filtro3
```

```python
# Carga el archivo de Excel y usa la onceava columna como índice del DataFrame
df1 = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx", index_col=0)
df1
```

```python
#Paso 2: Aplicamos filtros de filas vs columnas. df1 necesita ser un DataFrame con índices, en este caso el índice es la columna 0.
#En este caso se filtra por las filas F37034 y F37042 y las columnas NOMBRE_VENDEDOR y se crea un nuevo archivo csv.
#Esto podria servir para obtener el nombre del vendedor de las facturas F37034 y F37042 o en caso contrario obtener el nombre del vendedor 
#de todas las facturas.
filtro4 = df1.loc[["F37034", "F37042"], ["NOMBRE_VENDEDOR"]]
filtro4.to_csv("filtroPorFilasColumnas.csv")
filtro4
```

```python
#Filtro por cabecera: obtiene las primeras 2 filas del DataFrame. Si no se especifica el número de filas, por defecto se obtienen las primeras 5 filas.
filtro5 = df.head(2)
filtro5.to_csv("filtroPorCabecera.csv")
filtro5
```

```python
#Filtro por cola: obtiene la ultima fila del DataFrame. En caso de no especificar el número de filas, por defecto se obtiene las ukltimas 5 filas.
filtro6 = df.tail(1)
filtro6.to_csv("filtroPorCola.csv")
filtro6
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtros por comparacion: En este caso se filtra por los productos que tengan un precio unitario mayor a 30000 y se crea un nuevo archivo csv.
filtro7 = df[df["PRECIO_UNITARIO"] > 30000]
filtro7.to_csv("filtroPorComparacion")
filtro7
```

**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#Filtro Y (AND): En este caso se filtra por los productos que tengan un precio unitario mayor a 1000 y que la fecha de la factura sea 2022-01-05.
filtro8 = df[(df["PRECIO_UNITARIO"] > 1000) & (df["FECHA_DOC"] == "2022-01-05")]
filtro8.to_csv("filtroPorAND.csv")
filtro8
```

```python
#Filtro o (OR): En este caso se filtra por los productos que tengan un precio unitario mayor a 6000 o que la fecha de la factura sea 2022-01-05.
filtro9 = df[(df["PRECIO_UNITARIO"] > 6000) | (df["FECHA_DOC"] == "2022-01-05")]
filtro9.to_csv("filtroPorOR.csv")
filtro9
```

```python
#Filtro not: En este caso se filtra por los productos que tengan un precio unitario menor o igual a 100.
filtro10=df[~(df["PRECIO_UNITARIO"] > 100)]
filtro10.to_csv("filtroPorNOT.csv")
filtro10
```
